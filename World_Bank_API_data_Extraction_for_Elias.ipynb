```python
import wbdata
import pandas as pd
from datetime import datetime

# Define country and time range
country = "SOM"
start_year = 1980
end_year = 2025

# Define indicators
indicators = {
    "FP.CPI.TOTL.ZG": "Inflation_CPI",
    "NE.CON.TOTL.ZS": "Aggregate_Demand_Final_Consumption_GDP",
    "NY.GDP.MKTP.KD": "Aggregate_Supply_GDP_Constant_USD",
    "PA.NUS.FCRF": "Exchange_Rate_LCU_per_USD"
}

# Fetch data
data = wbdata.get_dataframe(
    indicators,
    country=country,
    date=(datetime(start_year, 1, 1), datetime(end_year, 12, 31))
)

# Reset index
data.reset_index(inplace=True)

# Rename columns
data.rename(columns={"date": "Year"}, inplace=True)

# Sort by year
data["Year"] = data["Year"].astype(int)
data = data.sort_values("Year")

# Save to CSV
data.to_csv("Somalia_WDI_Inflation_Data.csv", index=False)

# Save to Excel
data.to_excel("Somalia_WDI_Inflation_Data.xlsx", index=False)

data
```
